```python
for i in rawstif['LIBELLE_ARRET']:
    frozenset(i) 
group_by_stop=rawstif.groupby('LIBELLE_ARRET')
best_stop=group_by_stop.agg({'NB_VALD':'sum'}).sort_values(by='NB_VALD',ascending=False)
print(best_stop.head(15))
print(best_stop.tail(5))
```

```python
for i in rawstif['CATEGORIE_TITRE']:
    frozenset(i) 
group_by_titre=rawstif.groupby('CATEGORIE_TITRE')
best_titre=group_by_titre.agg({'NB_VALD':'sum'}).sort_values(by='NB_VALD',ascending=False)
print(best_titre.head(15))
```

# Création d'une table des validations par titre et par arrêt

```python
table = pd.pivot_table(rawstif, values='NB_VALD', index=['LIBELLE_ARRET'],columns=['CATEGORIE_TITRE'], aggfunc=np.sum)
table2 = pd.pivot_table(rawstif, values='NB_VALD', index=['ID_REFA_LDA'],columns=['CATEGORIE_TITRE'], aggfunc=np.sum)
table2['LDA_ID_REF_A']=table2.index
print(table2.tail(5))
```

```python
table2.fillna(0, inplace=True)
table2['etudiant']=table2['IMAGINE R'] / (table2['NAVIGO']+table2['IMAGINE R']+table2['TST']+table2['AMETHYSTE'])
table2['solidaire']=table2['TST']/(table2['NAVIGO']+table2['IMAGINE R']+table2['TST']+table2['AMETHYSTE'])
table2['ameth']=table2['AMETHYSTE']/(table2['NAVIGO']+table2['IMAGINE R']+table2['TST']+table2['AMETHYSTE'])

table2.head(2)
```

```python
arret=pd.read_pickle('arrets_stif')
arret.head(1)
```

```python
print(arret['LDA_ID_REF_A'].nunique())
print(arret.count())

table2.index.names = ['LDA_ID_REF_A']
arret=arret.sort_values(['LDA_ID_REF_A'])
arret=arret[arret.ZDEr_ID_TYPE_ARRET != 5]
arret.head(2)
print(arret.count())
```

```python
table3=pd.merge(left=table2, right = arret, on= 'LDA_ID_REF_A')
print(table3.dtypes)
print(table3['LDA_ID_REF_A'].nunique())
print(table3.count())
```

```python
print(table3['etudiant'].mean())
plt.hist(table3['etudiant'])

```

```python
plt.hist(table3['ameth'])
print(table3['LDA_ID_REF_A'].nunique())
print(table3.count())

```
